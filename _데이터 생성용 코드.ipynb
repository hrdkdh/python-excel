```python
import pandas as pd

df_raw = pd.read_clipboard()
df_raw
```

```python
df_raw.columns.to_list()
```

```python
df = df_raw.copy()
drop_cols = []
drop_strs = ["예산", "분관내용", "팩스", "전화번호", "홈페이지", "어린이", "노인", "도서관코드", "도서_", "도서자료_", "결산", "사서", "직원", "작은", "비도서", "향토", "전자자료", "연속", "설비", "지원인력", "도서관장"]
cols = df.columns
for col in cols:
    for drop_str in drop_strs:
        if drop_str in col:
            drop_cols.append(col)
        
df.drop(columns = drop_cols, inplace=True)
df.columns
```

```python
remain_col = ["도서관명", "주소", "상세주소", "설립주체_기관명", 
              "개관년도", "이동_장서수", "이동_이용자수", "시설_연면적", "시설_도서관부지면적",
              "좌석_총좌석수"]
```

```python
df_result = df.loc[(df["지역"] == "경북"), remain_col]
df_result["주소_"] = df_result["주소"] + df_result["상세주소"]
df_result.drop(columns = ["주소", "상세주소"], inplace=True)
df_result.rename(columns = { 
    "주소_" : "주소",
    "설립주체_기관명" : "설립주체",
    "이동_장서수" : "소장자료, 도서(권)",
    "이동_이용자수" : "이용자수(명))",
    "좌석_총좌석수" : "열람석(석)",
    "시설_연면적" : "연면적(m²)",
    "시설_도서관부지면적" : "대지(m²)"
}, inplace=True)
col_for_align = ["도서관명",
    "주소",
    "설립주체",
    "개관년도",
    "대지(m²)",
    "연면적(m²)",
    "열람석(석)",
    "소장자료, 도서(권)",
    "이용자수(명))"]
df_result =df_result[col_for_align]
df_result.reset_index(inplace=True)
df_result.drop(columns = ["index"], inplace=True)
```

```python
df_result["설립주체"] = df_result["설립주체"].str.replace("경상북도 포항시", "포항시")
```
