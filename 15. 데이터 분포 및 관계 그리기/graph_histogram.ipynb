```python
import pandas as pd
import matplotlib as mp
import matplotlib.pyplot as plt

#그래프 한글폰트 적용
mp.rc("font", family="Malgun Gothic")

df = pd.read_excel("graph_sample_data.xlsx", "학생건강검사 데이터", index_col="id")

#데이터 전체 히스토그램
print("데이터 전체 히스토그램")
df["height"].hist()
plt.show()

#고등학생만 필터링한 히스토그램
print("고등학생만 필터링한 히스토그램")
s_data = df.loc[df["sch_grade"] == "고",:]
s_data["height"].hist()
plt.show()

#고2만 필터링한 히스토그램
print("고2만 필터링한 히스토그램")
sg_data = s_data.loc[s_data["grade"] == 2,:]
sg_data["height"].hist()
plt.show()

#고2 중 성별로 구분한 히스토그램
print("고2 중 성별로 구분한 히스토그램")
sgsm_data = sg_data.loc[sg_data["sex"]=="남",:]
sgsf_data = sg_data.loc[sg_data["sex"]=="여",:]
sgsm_data["height"].hist()
sgsf_data["height"].hist()
plt.show()
```
